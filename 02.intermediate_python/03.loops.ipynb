## Introduction to Python Loops

Python loops are used to execute a block of code repeatedly until a certain condition is met. There are two types of loops in Python: while and for loops.

## While Loops
A `while` loop is used to repeat a block of code as long as the specified condition is True. The basic syntax of a `while` loop is as follows:

```python
while condition:
# code to execute while the condition is True
```

- Here, condition is a Boolean expression that is checked at the beginning of each iteration. If the condition is True, the code inside the loop is executed. This continues until the condition becomes False. Let's look at an example:

````python
# Simple while loop example
count = 0

while count < 5:
    print("Count is", count)
    count += 1
````

- In this example, the while loop prints the value of count as long as it is less than 5. The count variable is incremented by 1 at the end of each iteration.

## For Loops

A `for` loop is used to iterate over a sequence (e.g., a list, tuple, or string) and execute a block of code for each element in the sequence. The basic syntax of a for loop is as follows:

```python
for variable in sequence:
    # code to execute for each element in the sequence

```

Here, variable is a temporary variable that takes on the value of each element in the sequence. The code inside the loop is executed once for each element in the sequence. Let's look at an example:

````python
# Simple for loop example
numbers = [1, 2, 3, 4, 5]
for number in numbers:
    print(number)
````

````python
# A more complete use of for
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch historical stock data
ticker = "AAPL"
data = yf.download(ticker, start="1980-01-01")

# Example 1: for loop to calculate daily returns
returns = []
for i in range(1, len(data)):
    price_today = data["Close"][i]
    price_yesterday = data["Close"][i - 1]
    daily_return = (price_today - price_yesterday) / price_yesterday
    returns.append(daily_return)

# Plot daily returns
plt.plot(data.index[1:], returns)
plt.xlabel("Date")
plt.ylabel("Daily Returns")
plt.title("Daily Returns for {}".format(ticker))
plt.grid(True)
plt.show()
````

````python
# areas list
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# Change for loop to use enumerate()
for index, a in enumerate(areas) :
    print("room " + str(index)+ ": "  +str(a))
````

#### `iterrows()`

method in Python is used to iterate over the rows of a DataFrame. It returns an iterator object that yields a tuple of the row index and the row data. The row data is a Series object that contains the values for each column in the row.

Here is an example of how to use the `iterrows()` method:

````python
import pandas as pd

df = pd.DataFrame({"A": [1, 2, 3], 
                   "B": [4, 5, 6]})

for index, row in df.iterrows():
    print(index, row)
````

The `iterrows()` method is a powerful tool that can be used to iterate over the rows of a DataFrame. It can be used to perform a variety of tasks, such as:

- Calculating the sum of the values in each row
- Finding the maximum value in each row
- Filtering rows that meet certain criteria
- Iterating over the rows of a DataFrame in a for loop

````python
# Calculate the sum of the values in each row
totals = []
for index, row in df.iterrows():
    totals.append(row.sum())

print(totals)
````

````python
# Find the maximum value in each row
max_values = []
for index, row in df.iterrows():
    max_values.append(row.max())

print(max_values)
````

````python
# Filter rows that meet certain criteria
filtered_df = df[df["A"] > 2]

for index, row in filtered_df.iterrows():
    print(index, row)
````

````python
# Iterate over the rows of a DataFrame in a for loop
for index, row in df.iterrows():
    print(index, row["A"], row["B"])
````

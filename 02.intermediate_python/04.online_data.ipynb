##  Online Dataset

- Requirement: `conda install erddapy`

```python
from erddapy import ERDDAP

e = ERDDAP(
    server= "https://erddap.bco-dmo.org/erddap/",
    protocol="tabledap",
    response="csv",
)
```

```python
e.dataset_id = "bcodmo_dataset_3773"   
```

```python
# Print the URL - check
url = e.get_download_url()
print(url)
```

```python
df = e.to_pandas(index_col="time (UTC)",parse_dates=True)
```

```python
df.head()
```

```python
# Set display options to show all rows and columns
pd.set_option('display.max_rows', None)
print(df.dtypes)
```

```python
pd.set_option('display.max_rows', 10)
```

```python
CTDTMP_average = df.groupby(df.index.year)['CTDTMP (degrees celsius)'].mean().reset_index()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,5)) 
plt.plot(CTDTMP_average["time (UTC)"],CTDTMP_average["CTDTMP (degrees celsius)"],c='red',marker='.',linestyle='-') 

plt.ylabel('degrees celsius')
plt.legend()
plt.yticks(rotation=90)

```

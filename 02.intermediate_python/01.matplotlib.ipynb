## Overview

Matplotlib is built around two main components: the Figure and the Axes.

- The Figure is the top-level container that holds everything together. It can contain one or more Axes.

- The Axes is the area on which the data is plotted. It contains two or three Axis objects (depending on whether a 2D or 3D plot is being created) that define the data limits.

## Line Plot:



```python
import numpy as np

# To use matlplotlib in your python script, import pyplot
import matplotlib.pyplot as plt

# Generate some random data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure and axes object
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels and a title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Sine Wave')

# Save the plot to a file
fig.savefig('sine_wave.png')
```

In this example, we first generate some random data using NumPy's `linspace()` function and the `sin()` function. We then create a figure and axes object using `plt.subplots()`. We call the `plot()` function on the axes object to create the line plot, and then use the `set_xlabel()`, `set_ylabel()`, and `set_title()` functions to add labels and a title to the plot. Finally, we save the plot to a file using `savefig()`.

### Load financial data from Yahoo

```python
# Load pandas 
import pandas as pd

url = "https://query1.finance.yahoo.com/v7/finance/download/%5EIXIC?period1=34560000&period2=1683244800&interval=1d&events=history&includeAdjustedClose=true"
financial_data = pd.read_csv(url, parse_dates=['Date'], index_col='Date')
```

### Plotting the data

Create a simple line plot of the S&P 500 closing prices:



```python
plt.plot(financial_data['Close'])
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('NASDAQ Closing Prices')
plt.show()
```
